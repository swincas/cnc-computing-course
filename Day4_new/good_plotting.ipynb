```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a1 = np.arange(0, 10, 0.1)
a2 = np.linspace(0, 20, 100)
a3 = np.array([-3., 4.3, 5.1, 24.7, 9.4, 13.2, 2.3, 7.4, 6.5, 8.9, 33.5, 46.7, 89.0, 90.7, 43.45])
a3_y = np.array([0.1, 7.4, 12.3, 19.2, 2.4, 3.6, 7.7, 21.3, 23.4, 5.0, 1.1, 6.7, 9.2, 7.7, 13.1])
a3_err = np.full_like(a3, 3.)
```

## Default

```python
plt.plot(a1, label='a1')
plt.plot(a2, label='a2')
plt.errorbar(a3, a3_y, yerr=a3_err, fmt='o', label='a3 with error')
plt.xlabel('number')
plt.legend()
plt.show()
```

## Making use of the Artists
(a little at least)

```python
fig, ax = plt.subplots(nrows=1, ncols=1)

ax.plot(a1, label='Test 1', c='r')
ax.plot(a2, label='Test 2')
ax.errorbar(a3, a3_y, yerr=a3_err, fmt='^',
            capsize=3, linewidth=1, markersize=5,
            label='Test 3 with error')
ax.set(xlabel='numbers')
ax.legend()

plt.show()
```

## Using rcParams

```python
from cycler import cycler
plt.rcParams['figure.dpi'] = 200 # set dots per inch (png figure resolution)
plt.rcParams['font.family'] = 'palatino' # set a font of your choice
plt.rcParams['axes.labelsize'] = 14 # set the default label size
plt.rcParams['xtick.labelsize'] = 10 # set the default tick label size
plt.rcParams['ytick.labelsize'] = 10 
plt.rcParams['axes.prop_cycle'] = cycler(color=['b', 'r', 'g', 'k', 'y']) # set the automatic colour cycle
%matplotlib inline
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(a1, label='Test 1')
ax.plot(a2, label='Test 2')
ax.errorbar(a3, a3_y, yerr=a3_err, fmt='^', c='k',
            capsize=3, linewidth=1, markersize=5,
            label='Test 3 with error')
ax.legend()
plt.show()
```

```python
# reset the parameters
plt.rcParams.update(plt.rcParamsDefault)
```
